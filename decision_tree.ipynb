```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import sys

if 'google.colab' in sys.modules:
  
  !pip install -q dtreeviz
import sys
import os
# add library module to PYTHONPATH
sys.path.append(f"{os.getcwd()}/../")
from sklearn.datasets import *
from dtreeviz.trees import *
from IPython.display import Image, display_svg, SVG
```

```python
iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target
tree_clf = DecisionTreeClassifier(max_depth=2)
tree_clf.fit(X, y)
```

```python
iris
```

We can visualize our decision tree


```python
viz = dtreeviz(tree_clf,
               X,
               y,
               target_name='size',  # this name will be displayed at the leaf node
               feature_names=iris.feature_names,
               title="Iris data set",
               fontname="Arial",
               title_fontsize=16,
               colors = {"title":"grey"}
              )
viz
```

```python
X_data = iris.data[np.random.randint(0, len(iris.data)),:]

viz = dtreeviz(tree_clf,
               X,
               y,
               target_name='progr',  # this name will be displayed at the leaf node
               feature_names=iris.feature_names,
               X=X_data, 
               show_node_labels = True
              )
viz
```
